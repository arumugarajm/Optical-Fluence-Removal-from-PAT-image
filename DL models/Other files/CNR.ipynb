```python
def CNR(I1,I2):
    mean1 = np.mean(I1)
    mean2 = np.mean(I2)
    std1 = np.std(I1)
    std2 = np.std(I2)
    cnr = abs(mean1 - mean2)/math.sqrt(((std1*std1)+(std2*std2)))
    return cnr
```
